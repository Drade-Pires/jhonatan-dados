```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Carregar os dados do arquivo CSV (ajuste o caminho conforme sua estrutura)
df = pd.read_csv('../dados/vendas.csv')

# Converter a coluna 'Data' para datetime
df['Data'] = pd.to_datetime(df['Data'])

# Filtrar intervalo de datas (exemplo: janeiro de 2022)
inicio = '2022-01-01'
fim = '2022-01-31'
df_filtrado = df[(df['Data'] >= inicio) & (df['Data'] <= fim)]

# Agrupar por data e produto, somando vendas diárias
vendas_diarias = df_filtrado.groupby(['Data', 'Produto'])['Vendas'].sum().reset_index()

# Configurar estilo do gráfico
sns.set_style('whitegrid')
plt.figure(figsize=(12, 6))

# Plotar gráfico de linha para vendas por data, separado por produto
sns.lineplot(data=vendas_diarias, x='Data', y='Vendas', hue='Produto', marker='o')

# Ajustar título e legendas
plt.title('Vendas diárias por produto (Jan 2022)', fontsize=16)
plt.xlabel('Data', fontsize=14)
plt.ylabel('Vendas', fontsize=14)
plt.xticks(rotation=45)
plt.legend(title='Produto')

# Salvar gráfico na pasta imagens (certifique-se de que a pasta existe)
plt.tight_layout()
plt.savefig('imagens/grafico_vendas_temporal.png')
plt.show()


```
